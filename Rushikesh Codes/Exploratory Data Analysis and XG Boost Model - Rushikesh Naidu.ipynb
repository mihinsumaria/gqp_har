Dropping the output variable

```python
df3=train.drop('activity', 1)
df3.corr()
```

```python
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i, j] >= threshold:
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    return colname # deleting the column from the dataset

    
```

**Random Forest**

We need to convert the data into numeric form by applying one hot encoding, as Random Forest (Scikit-Learn Implementation) takes only numeric inputs. 

```python
train_X = train.drop(['activity'], axis=1)
train_y = train['activity']
test_X = test.drop(['activity'], axis=1)
test_y = test['activity']
```

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
# random forest model creation
rfc = RandomForestClassifier()
rfc.fit(train_X,train_y)
# predictions
rfc_predict = rfc.predict(test_X)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix

rfc_cv_score = cross_val_score(rfc, train_X, train_y, cv=10)
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(test_y, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(test_y, rfc_predict))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```

```python
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score
cm = confusion_matrix(test_y, rfc_predict) 
cm_df = pd.DataFrame(cm,
                     index = ['STANDING', 'STAND_TO_SIT', 'SITTING', 'SIT_TO_STAND', 'STAND_TO_LIE'
 ,'LAYING', 'LIE_TO_SIT', 'SIT_TO_LIE', 'LIE_TO_STAND', 'Walking',
 'WALKING_DOWNSTAIRS', 'WALKING_UPSTAIRS'], 
                     columns = ['STANDING', 'STAND_TO_SIT', 'SITTING' ,'SIT_TO_STAND', 'STAND_TO_LIE',
 'LAYING', 'LIE_TO_SIT', 'SIT_TO_LIE', 'LIE_TO_STAND', 'Walking',
 'WALKING_DOWNSTAIRS', 'WALKING_UPSTAIRS'])

plt.figure(figsize=(20,10))
sns.heatmap(cm_df, annot=True)
plt.title('Random Forest \nAccuracy:{0:.3f}'.format(accuracy_score(test_y, rfc_predict)))
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
features = train_X.columns
importances = rfc.feature_importances_
indices = np.argsort(importances)[-19:]  # top 20 features
plt.figure(figsize=(7,7))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')

plt.show()
```

```python
import time
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=100)
training_start = time.perf_counter()
xgb.fit(train_X, train_y)
training_end = time.perf_counter()
prediction_start = time.perf_counter()
preds = xgb.predict(test_X)
prediction_end = time.perf_counter()
acc_xgb = (preds == test_y).sum().astype(float) / len(preds)*100
xgb_train_time = training_end-training_start
xgb_prediction_time = prediction_end-prediction_start
print("XGBoost's prediction accuracy is: %3.2f" % (acc_xgb))
print("Time consumed for training: %4.3f" % (xgb_train_time))
print("Time consumed for prediction: %6.5f seconds" % (xgb_prediction_time))
```
